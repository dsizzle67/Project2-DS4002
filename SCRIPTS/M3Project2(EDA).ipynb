**Code Purpose:** This chunk of code plots Duolingo search interest over time
### **IMPORTANT:** In order for code to run you must upload time_series_US.csv

```python
import pandas as pd

# Load the CSV file into a DataFrame
search_data = pd.read_csv('time_series_US.csv')

# Convert the Time column to datetime objects
search_data['Time'] = pd.to_datetime(search_data['Time'])

# Set Time as the index for time‑series plotting
search_data.set_index('Time', inplace=True)

# Plot the Duolingo search interest over time
search_data['duolingo'].plot(figsize=(10, 6), title='Duolingo Interest Over Time')
```

**Code Purpose:** This code merges search‑interest and performance data, then plots search interest alongside DAU, subscribers, and revenue on dual axes to compare how they evolve over time.
### **IMPORTANT:** In order for code to run you must download activity_data_final.csv and dau_data_final.csv

```python
import matplotlib.pyplot as plt

# Create the figure and primary axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Load search‑interest and revenue/usage datasets
search = pd.read_csv('activity_data_final.csv')
revenue = pd.read_csv('dau_data_final.csv')

# Merge both datasets on the shared 'Period' column
df_merged = pd.merge(search, revenue, on='Period')

# Plot search interest on the primary y-axis
ax1.plot(df_merged['Period'], df_merged['duolingo'], marker='o')
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Search Interest ("duolingo")')
ax1.tick_params(axis='x', rotation=45)

# Create a secondary y-axis for performance metrics
ax2 = ax1.twinx()
ax2.plot(df_merged['Period'], df_merged['DAU'], linestyle='--', marker='s', label='DAU')
ax2.plot(df_merged['Period'], df_merged['Paid_Subscribers'], linestyle='--', marker='^', label='Subscribers')
ax2.plot(df_merged['Period'], df_merged['Revenue'], linestyle='--', marker='d', label='Revenue')
ax2.set_ylabel('Performance Metrics')

# Add legend for the secondary axis
ax2.legend(loc='upper left')

# Title and layout
plt.title('Trends in Search Interest and Performance')
plt.tight_layout()
plt.show()


```
